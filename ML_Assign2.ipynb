```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("/content/Heart (1).csv")
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
print(dataset.isnull().sum())
#Step 1: Handle missing values
```

```python
dataset['Ca']=dataset['Ca'].fillna(dataset['Ca'].mean())
#Fill missing values in Ca with mean
```

```python
dataset['Thal']=dataset['Thal'].fillna(dataset['Thal'].mode()[0])
#Fill missing values in Thal with mode
```

```python
print(dataset.isnull().sum())
#Step 1: Handle missing values
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2, figsize=(14,6))

sns.histplot(ax=axes[0],x=dataset.Age)
axes[0].set_title("Distribution of Age")

sns.histplot(ax=axes[1], x=dataset.Age, y=dataset.Chol)
axes[1].set_title(" Cholestrol among the ages")
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2, figsize=(14,6))

sns.histplot(ax=axes[0],x=dataset.Sex)
axes[0].set_title("Distribution of Sex")

sns.histplot(ax=axes[1], x=dataset.Age, y=dataset.RestBP)
axes[1].set_title(" RestBP among the ages")
```
