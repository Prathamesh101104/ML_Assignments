```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/admission.csv")
data.head()
```

```python
data.describe()
```

```python
sns.countplot(x='University Rating',data=data)
plt.title('Distribution of University Rating')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 7))
plt.pie(data['Research'].value_counts(), labels=data['Research'].value_counts().index, autopct='%1.1f%%')
plt.title('Distribution of Research Done by Students ')
plt.show()
```

```python
corr_matrix=data.corr()
corr_matrix
```

```python
corr_matrix=data.corr()
plt.figure(figsize=(15,10))
sns.heatmap(corr_matrix,annot=True,cmap='YlGnBu',linewidth=0.5,fmt=".2f")
plt.title("Correlation matrix of admission dataset")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data.columns = data.columns.str.strip()

sns.set(style='whitegrid')

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Scatter Plot: GRE Score vs. Chance of Admit
sns.scatterplot(x=data['GRE Score'], y=data['Chance of Admit'], alpha=0.6, ax=axes[0], color='b')
axes[0].set_title('GRE Score vs. Chance of Admit')
axes[0].set_xlabel('GRE Score')
axes[0].set_ylabel('Chance of Admit')

# Scatter Plot: TOEFL Score vs. Chance of Admit
sns.scatterplot(x=data['TOEFL Score'], y=data['Chance of Admit'], alpha=0.6, ax=axes[1], color='g')
axes[1].set_title('TOEFL Score vs. Chance of Admit')
axes[1].set_xlabel('TOEFL Score')
axes[1].set_ylabel('Chance of Admit')

# Scatter Plot: CGPA vs. Chance of Admit
sns.scatterplot(x=data['CGPA'], y=data['Chance of Admit'], alpha=0.6, ax=axes[2], color='r')
axes[2].set_title('CGPA vs. Chance of Admit')
axes[2].set_xlabel('CGPA')
axes[2].set_ylabel('Chance of Admit')

# Show plots
plt.tight_layout()
plt.show()
```

```python
#bar plot
plt.bar(data['University Rating'], data['Chance of Admit'], color='purple')
plt.xlabel('University Rating')
plt.ylabel('Chance of admit')
plt.title('University Rating VS Chance of admit')
plt.show()
```
